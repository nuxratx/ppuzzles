```python
from random import choice
def select_word():
    with open ("words.txt", mode= "r") as words: #In python to work with files,you have to open the file first 
        words_list = words.readlines()
        co = choice(words_list).strip()
        return co
    
        #This function simply takes the text file, saves as as words and then reads in lines, returning a random word
        
def get_player_input(guessed_letters):
    players_input_str = " "
    while len(players_input_str) != 1 or not players_input_str[0].isalpha or players_input_str in guessed_letters:
        #This is a compund while & if statement 
        players_input_str = input ("Guess a Letter: ").lower() 
        # .lower function ensure that the word is made into lower case
    return players_input_str
    
def join_guessed_letters(guessed_letters):
    guessed = " ".join(guessed_letters)
    return guessed 


def build_guessed_word(target_word, guessed_letters):
    current_list = []
    for i in target_word:
        res = '_'
        if i in guessed_letters:
            res = i
        current_list.append(res)
    return "".join(current_list)

def draw_hanged_man(wrong_guesses):
    hanged_man = [
        r"""
  -----
  |   |
      |
      |
      |
      |
      |
      |
      |
      |
-------
""",
        r"""
  -----
  |   |
  O   |
      |
      |
      |
      |
      |
      |
      |
-------
""",
        r"""
  -----
  |   |
  O   |
 ---  |
  |   |
  |   |
      |
      |
      |
      |
-------
""",
        r"""
  -----
  |   |
  O   |
 ---  |
/ |   |
  |   |
      |
      |
      |
      |
-------
""",
        r"""
  -----
  |   |
  O   |
 ---  |
/ | \ |
  |   |
      |
      |
      |
      |
-------
""",
        r"""
  -----
  |   |
  O   |
 ---  |
/ | \ |
  |   |
 ---  |
/     |
|     |
      |
-------
""",
        r"""
  -----
  |   |
  O   |
 ---  |
/ | \ |
  |   |
 ---  |
/   \ |
|   | |
      |
-------
""",
    ]

    print(hanged_man[wrong_guesses])
    
    
def game_over(wrong_guesses, target_word, guessed_letters):
        filtered_word= "".join([letter for letter in target_word if letter in guessed_letters])
        if wrong_guesses >= 6 or filtered_word == target_word:
            return True
        return False 
```

```python
#intial_game_setup
print()
word_to_guess = select_word()
input_letters = []
mystery_word = build_guessed_word(word_to_guess, input_letters)
wrong_guesses = 0
#Main Game Loop control to utilize all the functions that we have created so far. 
while not game_over(wrong_guesses, word_to_guess, input_letters):
    print(mystery_word)
    draw_hanged_man(wrong_guesses)
    print(join_guessed_letters(input_letters))
    input_letters.append(get_player_input(input_letters))
    mystery_word = build_guessed_word(word_to_guess, input_letters)
    if input_letters[-1] not in word_to_guess: 
        wrong_guesses += 1 # b = b + 1 or b += 1
print(mystery_word)
draw_hanged_man(wrong_guesses)
if mystery_word == word_to_guess:
    print("You got it!")
else:
    print("The word was ", word_to_guess)


    
```
